```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x = np.linspace(0,10,1000)

y_exact = x*(1+x)**-2.5
y_approx = stats.norm(loc=0.69, scale=1).pdf(x)

plt.figure()
plt.plot(x, y_exact, label="Gelman+ 2014")
plt.plot(x, y_approx, label="Nagpal+ 2022")
plt.legend(fontsize=16)
plt.xlabel("a,b", fontsize=20)
plt.yticks([])
plt.show()
```
